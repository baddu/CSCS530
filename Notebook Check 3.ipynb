```python
%matplotlib inline 

# for plotting in the notebook itself....rather than in a seperate window
import matplotlib.pyplot as plt
# Create graph
import networkx

import seaborn; seaborn.set()

# Import widget methods
from IPython.html.widgets import *

g = networkx.Graph()

num_people = 10
people = ['l','l','l','h','h','h','h','h','h','h']

g.add_nodes_from(range(num_people))
#g.add_edges_from(edge_weight.keys())
g_layout = networkx.spring_layout(g, iterations=100)

# Get node info
healthy_nodes = [i for i in range(num_people) if people[i] == 'l']
infected_nodes = [i for i in range(num_people) if people[i] == 'h']

# Now we can visualize the infection network.
f = plt.figure(figsize=(16, 16))
ax = f.gca()
networkx.draw_networkx_nodes(g, g_layout,
                       nodelist=infected_nodes,
                       node_size=25,
                       node_color='#b92732')

networkx.draw_networkx_nodes(g, g_layout,
                       node_size=25,
                       nodelist=healthy_nodes,
                       node_color='#64a171')

#
cmap = seaborn.cubehelix_palette(light=1, as_cmap=True)

networkx.draw_networkx_edges(g, g_layout, 
                       width=1, 
                       alpha=0.5,
                       edge_cmap=cmap,
                       edge_color='#64a171')

ax.grid(False)
ax.set_axis_bgcolor("#111111")
```
